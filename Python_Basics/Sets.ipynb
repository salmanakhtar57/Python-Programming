```python
#sets using intersection
S = {'A', 'B', 'C'}

U = {'A', 'Z', 'C'}

U.intersection(S)
```

```python
#Disjoint functions
s.add(1)
s.add(2)
s1 = {4, 6}
print(s.isdisjoint(s1))
```

```python

#Remove Function
s.add(1)
s.add(2)
s.remove(2)
print(s)
```
