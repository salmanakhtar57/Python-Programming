```python
#Here are some of the if-else examples that I found interesting. 
#if-else code using boolean function True or False with return function.

is_Python = False
is_Web = True
if is_Python and is_Web:
    print("You will get a Job at Lahore.")
elif is_Python and not(is_Web):
    print("You will go to Islamabad for more learning.")
elif not(is_Python) and (is_Web):
    print("You will do Freelaucing.")
else:
    print("You are wasting you TIME.")
```

```python
#if else with == operator.

def max_num(num1, num2, num3):
    if num1>=num2 and num1>=num3:
        return num1
    elif num2>=num1 and num2>=num3:
        return num2
    else:
        return num3
    
print(max_num(54123, 14300000, 513514))
```

```python
#Task
#Given an integer, , perform the following conditional actions:

#If  is odd, print Weird
#If  is even and in the inclusive range of  to , print Not Weird
#If  is even and in the inclusive range of  to , print Weird
#If  is even and greater than , print Not Weird

n = int(input())
if n % 2 == 0:
    if n in range(2,  6):
        print("Not Weird")
    elif n in range(6, 21):
        print("Weird")
    elif n > 20:
        print("Not Weird")
else:
    print("Weird") 
```
