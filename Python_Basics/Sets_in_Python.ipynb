```python
a = set()
print(a)
```

```python
a.add(3)
print(a)
```

```python
a.add(123)
print(a)
```

```python
for x in a:
  print(x)
```

```python
given_list = [1, 1, 2, 4, 2]

new_set = set()
for x in given_list:
  new_set.add(x)
print(new_set)
```

```python
#for list
new_list = list()
for x in new_set:
  new_list.append(x)
print(new_list)
```

```python
#sets does not necerrally follow the orders of elements.

b = set()
b.add('Pakistan')
b.add('UK')
b.add(512)
print(b)
```

```python
#finding the sum of unique elements in the given list.
given_list2 = [1, 3, 3, 1, 4, 5]
```

```python
new_set2 = set()
for x in given_list2:
  new_set2.add(x)
  total = 0
for x in new_set2:
  total += x
print(total)
```
